```python
import json

with open('../clothing/user_infos.json', 'r') as f:
    data = json.load(f)
```

```python
with open('../clothing/meta_dict.json', 'r') as f:
    meta = json.load(f)


with open('../clothing/interactions.json', 'r') as f:
    interactions = json.load(f)
```

```python
bad_id = []

for key, val in data.items():
    try:
        target_items = list(val['target_needs'].keys())
        val['targets'] = target_items
        val['target_category'] = meta[target_items[0]]['categories']
        val['target_price'] = interactions[key]['target_price']
    except:
        bad_id.append(key)
        
bad_id
```
